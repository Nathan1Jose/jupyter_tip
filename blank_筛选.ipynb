```python
 # 转换为灰度图像\
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 使用Otsu算法把灰度图的像素点分为两类，白0或黑255，thresh是二值化后的图像
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
# 查找图像中的轮廓
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# 导入PIL库，用于图像处理
from PIL import Image

# 定义一个函数，用于获取图像的像素值
def get_pixel_values(image):
    # 打开图像
    img = Image.open(image)
    # 转换为灰度图像
    gray = img.convert('L')
    # 获取图像的宽度和高度
    width, height = img.size
    # 创建一个空列表，用于存储像素值
    pixel_values = []
    # 遍历图像的每个像素
    for x in range(width):
        for y in range(height):
            # 获取像素的灰度值
            pixel = gray.getpixel((x, y))
            # 将像素值添加到列表中
            pixel_values.append(pixel)
    # 返回像素值列表
    return pixel_values

# 定义一个函数，用于获取列表中出现次数最多的前n个元素
def max_common(lst, n):
    # 创建一个字典，用于存储元素及其出现次数
    count = {}
    # 遍历列表中的每个元素
    for item in lst:
        # 如果元素已经在字典中，增加其出现次数
        if item in count:
            count[item] += 1
        # 否则，将元素及其出现次数设为1
        else:
            count[item] = 1
    # 根据出现次数对字典进行降序排序，返回前n个元素及其出现次数的列表
    return sorted(count.items(), key=lambda x: x[1], reverse=True)[:n]

# 定义一个函数，用于判断图像是否存在大部分空白背景
def is_blank(image):
    # 获取图像的像素值
    pixel_values = get_pixel_values(image)
    # 获取像素值出现次数最多的前5个元素及其出现次数
    top5 = max_common(pixel_values, 5)
    # 遍历前5个元素
    for value, count in top5:
        # 如果元素的值大于200，即接近白色，说明图像存在大量空白背景
        if value > 200:
            # 返回True，表示图像是空白的
            return True
    # 否则，返回False，表示图像不是空白的
    return False

# 定义一个函数，用于对一个文件夹里的图片进行筛选，把空白过多的图像给过滤掉
def filter_blank_images(folder):
    # 导入os库，用于操作文件和文件夹
    import os
    # 获取文件夹中的所有文件名
    files = os.listdir(folder)
    # 创建一个空列表，用于存储空白的图片文件名
    blank_images = []
    # 遍历文件夹中的每个文件
    for file in files:
        # 获取文件的完整路径
        file_path = os.path.join(folder, file)
        # 判断文件是否是图片，可以根据文件的扩展名进行判断，例如.jpg, .png, .bmp等
        if file_path.endswith('.jpg') or file_path.endswith('.png') or file_path.endswith('.bmp'):
            # 判断图片是否是空白的
            if is_blank(file_path):
                # 如果是空白的，将图片文件名添加到列表中
                blank_images.append(file)
    # 返回空白图片文件名的列表
    return blank_images

# 测试代码
# 假设有一个名为images的文件夹，里面存放了一些图片
folder = r'D:\desk\4th  space\Work\search\image\split'
# 调用函数，对文件夹里的图片进行筛选，把空白过多的图像给过滤掉
blank_images = filter_blank_images(folder)
# 打印结果
print('以下是空白过多的图像：')
for image in blank_images:
    print(image)
```

```python
# 打开一个名为output.txt的文件，如果不存在则创建，如果存在则覆盖
f = open(r'D:\desk\4th  space\Work\search\output.txt', 'w')
# 把列表转换为字符串，使用逗号分隔每个元素
s = ','.join(blank_images)
# 把字符串写入文件中
f.write(s)
# 关闭文件
f.close()
```
