```python
from PIL import Image
Image.MAX_IMAGE_PIXELS = None#解除像素限制
```

```python
img = Image.open(r'D:\desk\4th  space\Work\search\image\gdc_download_20231105_062048.057377\1f18d8dd-13e5-40ef-acd7-2c5591e7f982\TCGA-EM-A2P0-01Z-00-DX2.C464E7F1-67AC-4069-B0F0-D01BFEE3B71D.svs')#读入文件
```

```python
# 获取图片的宽度和高度
width, height = img.size
# 计算需要分割的行数和列数
rows = height // 224
cols = width // 224
print(width,height,rows,cols)
```

```python
# 遍历每个小块
for i in range(rows):
    for j in range(cols):
        # 计算每个小块的左上角和右下角的坐标
        left = j * 224
        top = i * 224
        right = (j + 1) * 224
        bottom = (i + 1) * 224
        # 裁剪图片
        img_crop = img.crop((left, top, right, bottom))
        # 保存图片，文件名为原始图片的文件名加上小块的编号
        img_crop.save(r'D:\desk\4th  space\Work\search\image\split\example_' + str(i) + '_' + str(j) + '.jpg')
```
